```python
# lets import os module to chek the current working diecroty and immport cars.csv file
import os

#define file name

srcFileName='cars.csv'

# lets find current worlking directory

curWrkDir = os.getcwd()

#print current value curWrkDir
print(curWrkDir)
```

```python
# lets find out a file with trgtrgtFileName and find it's absoulute path

for root, dir, files in os.walk(curWrkDir):
    for name in files:
        if name == srcFileName:
            finalPath = os.path.abspath(os.path.join(root,srcFileName))
            print(finalPath)
    
```
