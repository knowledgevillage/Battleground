# Make it Easy : TAB Completion and Tooltip

```python
import pandas as pd
```

```python
#import libraries
import os

cWrkDir=os.getcwd()
print(cWrkDir)
```

```python
# lets set fileName which we have to read
rawFile='titanic.csv'
```

```python
# lets search for all folders and sub folder under current directory to get the file path

for root, dir, files in os.walk(cWrkDir):
    for name in files:
        if name == rawFile:
            finalPath=os.path.abspath(os.path.join(root,name))
            print(finalPath)
```

```python
# lets read the file

df_titanic=pd.read_csv(finalPath)
```
