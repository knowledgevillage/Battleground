### search for dataset path

```python
#source filename
srcFileName='international_matches.csv'
```

```python
# lets serch for this file under current working directory

curWrkDir=os.getcwd()

for root, dir, files in os.walk(curWrkDir):
    for name in files:
        if name == srcFileName:
            finalPath=os.path.abspath(os.path.join(root,name))
            print(finalPath)
```

```python
#let's read csv file
df_Fifa=pd.read_csv(finalPath)
```

```python
# let's create a dataframe for team with rank

df_FifaRank= df_Fifa[['date', 'home_team', 'home_team_fifa_rank', 'home_team_total_fifa_points', 'away_team', 'away_team_fifa_rank', 'away_team_total_fifa_points']]


```

```python
df_FifaHome=df_FifaRank[['date','home_team', 'home_team_fifa_rank', 'home_team_total_fifa_points']].rename(columns={'home_team':'team', 'home_team_fifa_rank':'rank', 'home_team_total_fifa_points': 'rank_points' })
df_FifaAway=df_FifaRank[['date','away_team', 'away_team_fifa_rank', 'away_team_total_fifa_points']].rename(columns={'away_team':'team', 'away_team_fifa_rank':'rank', 'away_team_total_fifa_points': 'rank_points' })
```

```python
df_FifaRank= df_FifaHome.append(df_FifaAway)
```
